反响传播的实质是:链式求导,通过计算损失函数对每个参数的梯度,来更新参数,从而最小化损失函数.

梯度是一层一层计算的,从后往前(从输出层到输入层),每一层的梯度都是下一层的梯度与当前层的参数的乘积.

清空梯度:在每次反向传播之前,需要清空之前的梯度,否则梯度会累加.

要进行完一次向前传播,才能进行一次反向传播.

---

不断的链式求导就容易造成梯度消失或梯度爆炸的问题
- 梯度消失(偏导数<1)
- 梯度爆炸(偏导数>1)
- 解决方法
  - 梯度裁剪(clip)
  - 重新初始化权重
  - 使用其他激活函数
    - ReLU
    - LeakyReLU
    - tanh
    - sigmoid
    - softmax
    

Sigmoid 接近 0 或 1 时，梯度接近于 0，容易造成梯度消失问题

Tanh 以0为中心，输出范围为(-1, 1), 解决了Sigmoid的梯度消失问题

RelU 输出范围为(0, +∞), 解决了Tanh的梯度消失问题, 但是在输入为负数时, 梯度为0, 容易造成"死亡ReLU"问题



# Droupout

Droupout解决过拟合问题, 但是会导致模型训练速度变慢

 dropout如果每层都使用, 会导致模型训练速度变慢,且效果会下降

 可以只在首层和尾层使用dropout, 中间层不使用

# 单层单向

```py
import torch
import torch.nn as nn

#1.创建Rnn模型
#参数说明:
#input_size: 输入特征维度,可以理解为数据的特征数量
#hidden_size: 隐藏层维度,可以理解为隐藏状态的特征数量
#num_layers: RNN层数(如果设置多个层,前一个隐藏层的输入,一般设置1即可,默认也是1)
#batch_first: 如果为True,则输入输出张量的形状为(batch_size, seq_len, input_size)
#bidirectional: 是否为双向RNN
rnn_model=nn.RNN(input_size=10,hidden_size=20,
num_layers=1,batch_first=True,bidirectional=False)


2.调用模型
输入数据:
input_data=torch.randn(3,5,10)#[batch_size:批量大小,seq_len:序列长度,input_size:输入特征维度]
hx=torch.randn(1,3,10)#[ num_layers:隐藏层层数,batch_size:批量大小, hidden_size:隐藏层维度]

输出数据:
output,hn=rnn_model(input_data,hx)#output:[batch_size,seq_len,hidden_size]
print(output.shape)
print(hn.shape)

# Lasso回归与岭回归：直观理解

在带有正则项的线性回归中，通过在损失中加入惩罚项来控制模型复杂度：

Lasso（L1）与Ridge（L2）的目标函数分别为：

$$ \min_{\beta} \; \frac{1}{2n}\|y - X\beta\|_2^2 + \alpha \|\beta\|_1 $$

$$ \min_{\beta} \; \frac{1}{2n}\|y - X\beta\|_2^2 + \alpha \|\beta\|_2^2 $$

- L1 会将部分系数压到 0，产生稀疏解，具备特征选择能力。
- L2 会整体收缩系数，降低方差，提升稳定性但不做特征选择。
- 一般噪声较大、特征众多时优先考虑 Lasso；多重共线性强时优先 Ridge。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
X, y = make_regression(n_samples=120, n_features=12, n_informative=3, noise=12, random_state=42)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
lasso = Lasso(alpha=0.2, random_state=42, max_iter=10000)
ridge = Ridge(alpha=1.0, random_state=42)
lasso.fit(X_scaled, y)
ridge.fit(X_scaled, y)
y_lasso = lasso.predict(X_scaled)
y_ridge = ridge.predict(X_scaled)
print("Lasso R2:", r2_score(y, y_lasso))
print("Ridge R2:", r2_score(y, y_ridge))
print("Lasso零系数数量:", np.sum(lasso.coef_ == 0.0))
```

```python
idx = np.arange(X.shape[1])
w_lasso = lasso.coef_
w_ridge = ridge.coef_
plt.figure(figsize=(8, 4))
plt.bar(idx - 0.2, w_lasso, width=0.4, label="Lasso")
plt.bar(idx + 0.2, w_ridge, width=0.4, label="Ridge")
plt.xlabel("特征索引")
plt.ylabel("系数值")
plt.title("Lasso vs Ridge 系数对比")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

结论：
- Lasso 将不重要的特征系数直接压为 0，实现自动特征选择。
- Ridge 将所有特征系数向零收缩，减少过拟合，提高数值稳定性。
- 正则强度由超参数 $\alpha$ 控制，需通过验证集或交叉验证选择。

```python
# 导入库与准备示例数据
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'PingFang SC', 'Heiti SC', 'SimHei'] 
plt.rcParams['axes.unicode_minus'] = False 

# 设置随机种子以确保随机数可复现
np.random.seed(42) 

# 连续 x 轴数据（50 个点），用于 line/scatter
x = np.linspace(0, 10, 50) 

# 构造两组 y 数据：在正弦/余弦基础上叠加少量噪声
y1 = np.sin(x) + 0.1*np.random.randn(50) 
y2 = np.cos(x) + 0.1*np.random.randn(50) 

# 柱状图类别与对应值
cats = ['A','B','C','D','E'] 
vals = [12, 20, 8, 15, 10] 

# 直方图随机样本（正态分布），用于统计频次
samples = np.random.normal(loc=0.0, scale=1.0, size=500) 

# 饼图大小与标签
sizes = [30, 25, 20, 15, 10] 
labels = ['苹果','香蕉','葡萄','橙子','草莓']
```

```python
# 折线图（plot）：展示随 x 变化的两条曲线
plt.figure(figsize=(7,4)) 
# 第一条曲线：点标记为圆（marker='o'），颜色为 C0
plt.plot(x, y1, label='sin + noise', color='C0', marker='o') 
# 第二条曲线：虚线（linestyle='--'），颜色为 C1
plt.plot(x, y2, label='cos + noise', color='C1', linestyle='--') 
# 标题与坐标轴标签
plt.title('折线图') 
plt.xlabel('横轴') 
plt.ylabel('纵轴') 
# 网格与图例
plt.grid(True, alpha=0.3) 
plt.legend() 
plt.show()
```

```python
# 散点图（scatter）：离散点的分布与趋势
plt.figure(figsize=(7,4)) 
# s=40 控制点大小，color 设置颜色
plt.scatter(x, y1, label='points', color='C2', s=40) 
plt.title('散点图') 
plt.xlabel('横轴') 
plt.ylabel('纵轴') 
plt.grid(True, alpha=0.3) 
plt.legend() 
plt.show()
```

```python
# 柱状图（bar）：类别-数值关系可视化
plt.figure(figsize=(7,4)) 
# cats 为类别，vals 为对应值
plt.bar(cats, vals, color='C3') 
plt.title('柱状图') 
plt.xlabel('类别') 
plt.ylabel('数值') 
# 仅对 y 轴绘制网格，便于比较高度
plt.grid(axis='y', alpha=0.3) 
plt.show()
```

```python
# 直方图（hist）：数值分布的频次统计
plt.figure(figsize=(7,4)) 
# bins 控制箱数，edgecolor 提升清晰度
plt.hist(samples, bins=25, color='C4', edgecolor='white') 
plt.title('直方图') 
plt.xlabel('数值') 
plt.ylabel('频次') 
plt.grid(axis='y', alpha=0.3) 
plt.show()
```

```python
# 饼图（pie）：组成占比的直观展示
plt.figure(figsize=(6,6)) 
# autopct 显示百分比，startangle 设定起始角度，colors 使用调色板
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors) 
plt.title('饼图') 
# 设为正圆显示
plt.axis('equal') 
plt.show()
```

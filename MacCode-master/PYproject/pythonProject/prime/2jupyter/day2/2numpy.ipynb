#### 索引和切片

```python
arr12=np.arange(0,12).reshape(3,4)
print(arr12)
#取一行(下标为1的一行,也就是第二行)
print("取一行(下标为1的一行,也就是第二行):",arr12[1])
#取一列(下标为2的一列,也就是第三列)
print("取一列(下标为2的一列,也就是第三列):",arr12[:,2])
#取连续的行和列
print("取不连续的行和列:",arr12[0:2,0:2])
#取不连续的多行(不包括第2行和第3行)
print("取不连续的多行(不包括第2行和第3行):",arr12[[0,2],:])
#取不连续的多列(不包括第2列和第3列)
print("取不连续的多列(不包括第2列和第3列):",arr12[:,[0,3]])
#取从1开始的所有行
print("取从1开始的所有行:",arr12[1:,:])
#
# print("",arr12[:,:,2:])
```

#### 修改元素值

```python
arr13 = np.arange(24).reshape(4, 6)  # 示例形状 (4, 6)
print("原始数组:")
print(arr13)
print('-' * 30)

# 1. 修改单值：将第2行第3列的值改为 999
arr13[1, 2] = 999
print("将第2行第3列的值改为 999,修改单值后:")
print(arr13)
print('-' * 30)

# 2. 修改某一行值：将第3行整行改为 [100, 101, 102, 103, 104, 105]
arr13[2, :] = np.array([100, 101, 102, 103, 104, 105])
print("将第3行整行改为 [100, 101, 102, 103, 104, 105],修改第3行后:")
print(arr13)
print('-' * 30)

# 3. 修改某一列值：将第4列整列改为 200
arr13[:, 3] = 200
print("将第4列整列改为 200,修改第4列后:")
print(arr13)
print('-' * 30)

# 4. 修改某一块值：将左上角 2×2 块改为 0
arr13[0:2, 0:2] = 0
print("将左上角2×2块改为 0,修改左上角2×2块后:")
print(arr13)
print('-' * 30)

# 5. 修改多个不相邻值：将 (0,1)、(2,3)、(3,5) 位置的值分别改为 77、88、99
rows = [0, 2, 3]
cols = [1, 3, 5]
vals = [77, 88, 99]
arr13[rows, cols] = vals
print("将 (0,1)、(2,3)、(3,5) 位置的值分别改为 77、88、99,修改多个不相邻值后:")
print(arr13)
print('-' * 30)

# 6. 根据条件修改：将所有大于 150 的值改为 -1
arr13[arr13 > 150] = -1
print("条件修改（>150→-1）后:")
print(arr13)
print('-' * 30)

# 7. 三目运算：将小于 100 的元素变为原来的两倍，否则减 50
arr13 = np.where(arr13 < 100, arr13 * 2, False)
print("三目运算后:")
print(arr13)
print('-' * 30)
```

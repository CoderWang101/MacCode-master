### 4.2DataFrame索引

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(np.arange(12).reshape(3,4),index=['s1','s2','s3'],columns=['c1','c2','c3','c4'])
df
```

```python
print("原始数据:")
print(df)
print("*"*50)
#1.通过列索引获得取值
print("1.通过列索引获得取值:")
print(df['c1'])
print("*"*50)
#2.通过列索引获得多个取值
print("2.通过列索引获得多个取值:")
print(df[['c1','c2']])
print("*"*50)
#3.通过行索引获得取值
print("3.通过行索引获得取值:")
print(df.loc['s2'])
#4.通过行索引获得多个取值
print("4.通过行索引获得多个取值:")
print(df.loc[['s1','s2']])
```

```python
#5.通过位置索引获得取值
print("5.通过位置索引获得取值:")
print(df.iloc[0])
print("*"*50)

#6.通过位置索引获得多个取值
print("6.通过位置索引获得多个取值:")
print(df.iloc[[0,2]])
print("*"*50)

#7.获取具体位置取值
print("7.获取具体位置取值:")
print(df.iloc[1,2])
```

```python
#8.通过位置索引切片获取数据
print("8.通过位置索引切片获取数据:")
print(df.iloc[0:2,1:3])
print("*"*50)
#9.通过名字索引切片获取数据
print("9.通过名字索引切片获取数据:")
print(df.loc['s1':'s2','c1':'c2'])
print("*"*50)
#10.一次取多个不连续的行或列
print("10.一次取多个不连续的行或列:")
print(df.iloc[[0,2],[1,3]])#前面是行索引，后面是列索引
print("或")
print(df.loc[['s1','s2'],['c1','c2']])
```

### 4.3 检测重复值


```python
data={
    "name":["Alice","Bob","Charlie","David","Alice","Charlie"],
    "age":[25,30,35,40,25,35],
    "score":[80,85,90,95,80,85],
    "gender":["F","M","M","F","F","M"]
    
}
```

```python
df=pd.DataFrame(data,index=["s1","s2","s3","s4","s5","s6"],columns=["name","age","score","gender"])
df
```

```python
#1.检测重复值
print("1.检测重复值:")
print(df.duplicated())
print("*"*50)
print("True表示重复值")
```

```python
#2.指定列检测重复值
print("2.指定列检测重复值:")
print(df.duplicated(subset=["name","age"]))#根据name和age列检测重复值
print("*"*50)
print("True表示重复值")
```

```python
#3.删除重复值
print("3.删除重复值:")
print(df.drop_duplicates(inplace=True))#inplace=True表示在原数据上操作
print("*"*50)
print("删除重复值后:")
df
```

### 4.4 缺失值处理

```python
data={
    "name":["Alice","Bob","Charlie","David","Alice","Charlie"],
    "age":[25,30,35,40,25,np.nan],
    "score":[80,np.nan,90,95,80,np.nan],
    "gender":["F","M","M",np.nan,np.nan,np.nan]
}
df=pd.DataFrame(data,index=["s1","s2","s3","s4","s5","s6"],columns=["name","age","score","gender"])
df
```

```python
print("检测缺失值:")
print("True表示缺失值")
print(df.isnull())
print("*"*50)
#计算缺失值的数量
print(df.isnull().sum())
print("*"*50)
#每一行的缺失值数量
print(df.isnull().sum(axis=1))
print("*"*50)
#每一列的缺失值数量
print(df.isnull().sum(axis=0))
```

```python
#计算每一列的缺失值比例
print(df.isnull().mean(axis=0))
print("*"*50)

#计算每一行的缺失值比例
print(df.isnull().mean(axis=1))
```

```python
#每一列缺失率
print(df.isnull().sum()/len(df))
#每一行缺失率
print(df.isnull().sum(axis=1)/df.shape[1])
```

```python
# 删除缺失值
print(df)
```

```python
#2.删除所有缺失值的行或列(谨慎操作,可能会丢失大量数据)
print(df.dropna())
```

```python
#3.删除所含缺失值的列
print("删除所含缺失值的列")
print(df.dropna(axis=1))
print('-'*30)
#4.删除所有缺失值的行
print("删除所有缺失值的行")
print(df.dropna())
print('-'*30)

# 5.删除指定列缺失值的行
df_drop_by_col=df.dropna(subset=["gender"])
print("删除gender列缺失值的行")
print(df_drop_by_col)
```

```python
#填充缺失值
print(df)
#1.填充全局缺失值,填充缺失值为0
df_fill_na=df.fillna(0)
print("填充缺失值为0")
print(df_fill_na)
print('-'*30)

#2. 填充指定列缺失值
df_fill_na_col=df.fillna({"age":0,"score":0})
print("填充age列缺失值为0,score列缺失值为0")
print(df_fill_na_col)
print('-'*30)

#3.单独把某列取出来,填充缺失值为80
df["age"]=df["age"].fillna(80)
print("单独把age列取出来,填充age列缺失值为80")
print(df)
print('-'*30)
```

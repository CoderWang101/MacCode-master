```python
import warnings
warnings.filterwarnings("ignore")
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = datasets.load_iris()
#data为150*4的矩阵,取data的所有行和第3、4列;target为150*1的矩阵,取target的所有行
X = iris.data[:, 2:4]
y = iris.target

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Value': [65, 70, 75, 50, 55, 60, 75, 80, 85]
})

plt.figure(figsize=(8, 5))
sns.set_theme(style="whitegrid")  # 设置主题风格
sns.boxplot(x='Category', y='Value', data=data, 
            palette="husl",  # 使用husl颜色主题
            medianprops={"color": "black", "linewidth": 2},  # 调整中位数线颜色和宽度
            whiskerprops={"linewidth": 2})  # 调整 whisker 线宽度

plt.title('Compare the data of the three groups', fontsize=14, pad=20)  # 调整标题字体和间距
plt.xlabel('Category', fontsize=12)  # 设置x轴标签
plt.ylabel('Value', fontsize=12)  # 设置y轴标签
plt.xticks(fontsize=10)  # 调整x轴刻度字体大小
plt.yticks(fontsize=10)  # 调整y轴刻度字体大小plt.show()
```

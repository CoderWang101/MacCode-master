```python
import pandas as pd
rowdata={'电影名称':['无问西东','后来的我们','前任3','红海行动','唐人街探案','战狼2'],
        '打斗镜头':[1,5,12,108,112,115],
        '接吻镜头':[101,89,97,5,9,8],
        '电影类型':['爱情片','爱情片','爱情片','动作片','动作片','动作片']
        }
movie_data = pd.DataFrame(rowdata)
movie_data
```

```python
new_data = [24,67]
data = movie_data.iloc[:,1:3]
data
```

```python
(((data-new_data)**2).sum(1))**0.5
```

```python
dist = list((((data-new_data)**2).sum(1))**0.5)
dist
```

```python
dist_len = pd.DataFrame({'dist':dist,'labels':(movie_data.iloc[:,3])})
dist_len
```

```python
k=4
dr = dist_len.sort_values(by='dist')[:k]
dr
```

```python
re = dr.loc[:,'labels'].value_counts()
re
```

```python
re.index[0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
X = movie_data[['打斗镜头','接吻镜头']]
y = movie_data['电影类型']

knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X,y)

new_data=[[24,67]]
predict_type = knn.predict(new_data)
print(f'预测的新电影类型是：{predict_type[0]}')
```
